```python
#import pandas library for accessing csv files
import pandas as pd
```

```python
dataset_1 = pd.read_csv("insurance_pre.csv")
dataset_1.head()
```

```python
dataset_1.shape
```

```python
dataset = pd.get_dummies(dataset_1,drop_first = True)
dataset.head()
```

```python
# Assign independent variable
independent = dataset[['age','bmi','children','sex_male','smoker_yes']]
independent
```

```python
# Assign dependent Variable
dependent = dataset[['charges']]
dependent
```

```python
# split train and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size = 0.30,random_state = 0)
x_train,x_test,y_train,y_test
```

```python
from sklearn.ensemble import AdaBoosterRegressor
```

```python
# official websit adaboost regressor = https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostRegressor.html
from sklearn.ensemble import AdaBoostRegressor
regressor = AdaBoostRegressor(estimator=None, n_estimators=50, learning_rate=1.0, loss='linear', random_state=None, base_estimator='deprecated' )
```

```python
# Model Creating
regressor.fit(x_train,y_train)
```

```python
# predict the test model
y_predict = regressor.predict(x_test)
y_predict
```

```python
# Evaluate the model
from sklearn.metrics import r2_score 
r_score = r2_score(y_test,y_predict)
r_score
```
